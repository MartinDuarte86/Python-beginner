# Listas
## Conceptos Basicas lista
+ Es igual a los array y vectores 
+ Se pueden almacenar diferente valores 
+ Se puede expadir dinamicamente a√±adiendo nuevos elementos
+ Declaracion de lista "Nombre_lista=[A,B,.....,Z] " se separa por "," 

## Comandos Basicos
+ ***lista.append*** : Agrega elementos al final de la lista
+ ***lista.insert(indices,"A1")***: Agrego un valor en un indice determinado.
+ ***lista.extend(['B1','C1','D1','E1'])***: para agregar varios valores juntos a una lista
+ ***lista.index("Valor a buscar")***: sirve para saber el indice de un valor adentro de la lista, en el caso de haber dos valores iguales devuelve siempre el primero.
+ ***Valor in lista*** Sirve para valor si existe un valor determinado dentro de la lista, solo retorna un valor boolenO TRUE: Existe  O FALSE: No existe 
+ ***lista.remove("Valor")*** elimino elementos
+ ***lista.pop*** elimino el ultimo elemento de la lista

```python
############ insertar
lista.append ('J') 
lista.insert(1,"A1")
lista.extend(["B1","C1","D1","E1"])

############ Buscar
print(lista.index("B"))
print("B"  in lista)


############ ELiminar 
lista.remove("A")
lista.pop()
```

## Visualizar elementos por indices 
+ Tiene indice empieza con indice 0 se puede recorrer con indices positivos [1] o negativo [-1] pero en este caso se no existen el indice 0 y recorrer desde el final hacia adelante.

```python
lista =["A","B","C","D","E","F"] 
print(lista[:]) ##Muestro toda la lista
print(lista[2]) ##Recorre el indices desde adelante hacia atras 
print(lista[-2])##Recorre el indices desde atras hacia  adelante
print(lista[0:3])##Recorre el indices desde "0" hacia  "3"
print(lista[:3])##Recorre el indices hacia el "3" pero se incluye por default desde el "0"
print(lista[:3])##Recorre el indices hacia el "3" pero se incluye por default desde el "0"
print( lista[2:])##Recorre el indices hacia el "3" pero se incluye por default desde el "0"
print( lista[-2:-1])##Recorre el indices hacia el "-2" pero se incluye por default desde el "-1"
```
