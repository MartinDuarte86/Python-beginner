## Comandos basicas
+ ***len(tupla)*** saber cuento elementos tiene la tupla
+ ***tupla[indice]*** Acceder a un indice
+ ***list(tupla)*** Convierte la tupla en lista (Para corroborrar si esta bien echa la conversion el print deberia verse [,,,])
+ ***tuple(tupla)*** Convierte lista en tupla (Para corroborrar si esta bien echa la conversion el print deberia verse (,,,))
+ ***valor in tupla*** Buscar elementos ; Devuleve un valor boolenado ***true*** "Si encontro el valor" o ***false*** "No encontro el valor"
+ ***tupla.count(valor)*** : cuanto elemento hay dentro de una tupla con el valor a buscar  
+ ***tupla("UNICO_VALOR",)*** para Crear tupla unitarias o con un unico elemento se tiene que poner "," al final del valor de la tupla.


```python
tupla=("A",1,1986,1986) 
print(tupla[0])

### TUPLA A LISTA
lista=list(tupla)
print(lista)

### LISTA A TUPLA
tupla=tuple(lista)
print(tupla)


### Busco valor en la tupla
print( 1 in tupla)
print(tupla.count(1986))

###Cantidad de elemento de la tuplas 
print(len(tupla))

### Tupla unitaria 
tupla2 =("UNICO_VALOR",)
print(len(tupla2))


### empaquetado de tuplas
tupla3 = "A","B","C","D"
print(tupla3)


### Desempaquetado de tuplas
tupla3 = ("Martin",24,5,1986)
nombre,dia,mes,anio= tupla3
print(nombre)
print(mes)
print(dia)
print(anio)

```
